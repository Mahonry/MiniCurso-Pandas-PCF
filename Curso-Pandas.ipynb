# Funciones Generales y Tipos De Datos 

```python
#Cargamos Datos con el método pd.read_csv(liga)
df_meteorites = pd.read_csv('./datasets/Meteorite_Landings.csv')
```

```python
#El método .sample() nos permite visualizar algunos registros aleatorios
df_meteorites.sample(10)
```

```python
#Para borrar una columnas usamos el método .drop(), axis = 1 hace referencia a las columnas, axis = 0 a las filas
df_meteorites.drop(['fall'], axis=1, inplace=True)
```

# Group by, Agregaciones y Pivot Table (Tablas dinamicas)

```python
#Importamos Seaborn ya que contiene un dataset util
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.sample(2)
```

```python
#Queremos conocer el promedio por sexo
df.groupby('sex').mean()
```

```python
df.groupby('sex').sum()
```

```python
#Supongamos que son dolares y queremos hacer el cambio a pesos mexicanos
def mean_usd2mxn(x):
    y = np.mean(x)*20
    return y
```

```python
df.groupby('sex')[['total_bill']].apply(mean_usd2mxn)
```

```python
#Podemos hacer agregaciones por dos o mas columnas
df.groupby(['sex','time'])[['total_bill']].apply(mean_usd2mxn)
```

```python
#Tambien funciona con funciones lambda
df.groupby(['sex','time'])[['total_bill']].apply(lambda x: np.mean(x)*20)
```

```python
#Podemos sacar minimos o maximos
df.groupby(['sex','time'])[['total_bill','size']].aggregate([np.mean,np.max])
```

```python
df.groupby(['sex','time'])[['total_bill','size']].agg([np.mean,np.max])
```

```python
#Podemos definir agregaciones personalizadas por columna
dict_agg = {'tip':[min, max],'total_bill':[np.mean, np.std]}
dict_agg
```

```python
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

```python
#Las pivot table nos permiten hacer agregaciones por columnas e indices
```

```python
df['count'] = 1
```

```python
df.pivot_table(values='count', index='sex', columns='time', aggfunc = np.sum)
```

```python
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc = [np.sum, np.mean])
```

```python
df.pivot_table(values='total_bill', index=['sex','smoker'], columns='time', aggfunc = [np.sum, np.mean])
```

# Joins

![Joins](./images/sqljoin.jpeg)

```python
#Creamos los DF's de prueba
df_left = pd.DataFrame(
                      {'X': ['x0', 'x1', 'x2', 'x3'],
                        'W': ['w0', 'w1', 'w2', 'w3'], 
                        'Y': ['y0', 'y1', 'y2', 'y3'] },
                       index=[0,1,2,3])
df_left
```

```python
df_right = pd.DataFrame(
                       {'Z': ['z2', 'z3', 'z4', 'z5'],
                         'A': ['a2', 'a3', 'a4', 'a5'], 
                         'Y': ['y2', 'y3', 'y4', 'y5']},
                        index=[2,3,4,5])
df_right
```

```python
#Por default el parametro del join es un inner 
pd.merge(df_left, df_right, on = 'Y')
```

```python
#hacemos un left
pd.merge(df_left, df_right, how = 'left', on = 'Y')
```

```python
pd.merge(df_left, df_right, how = 'right', on = 'Y')
```
